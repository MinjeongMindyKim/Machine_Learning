```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
#층들을 쌓아주는, 하나의 층을 만들어줄 수 있는 객체
```

```python
import numpy as np
import tensorflow as tf
```

```python
np.random.seed(3)
tf.random.set_seed(3)
```

```python
Data_set = np.loadtxt("./dataset/ThoraricSurgery.csv", delimiter=",")
```

```python
type(Data_set)
```

```python
X = Data_set[:,0:17]
Y = Data_set[:, 17]
```

```python
model = Sequential()
#시퀀스객체: 통으로 레이어를 하나 깔아줌. 큰 틀
#전체적인 판을 하나 만들고 베이스를 만들기 시작. 그건 add를 사용해서 판에 넣음
model.add(Dense(30, input_dim=17, activation='relu')) #-일 경우 값을 0으로 만들어줌 / 은닉층 함수가 렐루
#x피쳐가 17개라는 것, 첫번째 히든레이어의 노드(동그라미)를 30개로 구성하겠다는 뜻
#(Dense(30, input_dim=17, activation='relu')): 열백터 하나를 만들어줌. 
# 인풋과 첫번째 히든 레이어를 같이 묶음. 그 다음은 한 층씩 만듦
model.add(Dense(1, activation='sigmoid')) #시그모이드는 이진분류에서 사용하는 것 / 노드 바로 전단계의 엑티베이트 펑션
#1은 마지막 덴스, 출력층. 인풋 딤은 전단계 출력 30개가 그 다음 덴스에 입력이 되는 것,
# 입력은 30개 출력은 1개다!

# 덴스가 세로로된 노드, 열백터처럼! 한층한층 쌓아서 시퀀셜 객체에 add해서 추가함
```

```python
model.compile(loss='binary_crossentropy', optimizer='adam' , metrics=['accuracy']) #컴파일은 일케 세팅해준다는 뜻
#이진분류라 바이너리 씀(카테고리컬 크로스엔트로피는 다중분류)
#w를 찾아가는데 방향과 사이즈를 모두 고려하는 아담을 쓰겠다
#매트리스는 사람이 보기 편한 평가 지표로 바꿔준다
model.fit(X, Y, epochs=100, batch_size=10)
#에포크: 딥러닝에 대한 프레임웤 노드를 만듦 / 모델을 바꾸면서 W를 바꾸겠다, 바꿔보겠다.
# 모델은 고정시켜놓고 데이터를 w를 바꿔서 측정하기 위해 데이터를 계속적으로 밀어넣는데, 이 모델을 몇 번 수행할 것이냐 하는 게 에포크
#데이터를 작게 잘라서 배치사이즈로 만든다. cpu성능에 부화가 안 되게. 배치사이즈를 10으로 나누겠다는 말. 총 470개면 47번 돌아감
```
