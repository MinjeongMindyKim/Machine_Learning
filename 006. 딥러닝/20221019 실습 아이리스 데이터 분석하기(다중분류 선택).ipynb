```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from sklearn.preprocessing import LabelEncoder
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
import tensorflow as tf
```

```python
np.random.seed(3)
tf.random.set_seed(3)
```

```python
df = pd.read_csv('./dataset/iris.csv', names = ["sepal_length", "sepal_width", \
                                              "patal_length", "petal_width", "species"])

print(df.head())
```

```python
sns.pairplot(df, hue='species');
#plt.show()
```

```python
dataset = df.values
X = dataset[:,0:4].astype(float)
Y_obj = dataset[:,4]
```

```python
X
```

```python
Y_obj
#아직 str. 레이블 인코더를 통해 레이블 인코딩을 해줘야 함
```

```python
e = LabelEncoder()
e.fit(Y_obj)
Y=e.transform(Y_obj)
Y
```

```python
Y_encoded = tf.keras.utils.to_categorical(Y)
Y_encoded
```

```python
#모델의 설정
model = Sequential()
model.add(Dense(16, input_dim=4, activation='relu'))
model.add(Dense(3, activation='softmax'))
```

```python
#모델 컴파일
model.compile(loss='categorical_crossentropy', #다중분류라 이 파라미터가 로스로 들어감
             optimizer='adam',
             metrics=['accuracy'])
```

```python
#모델 실행
model.fit(X, Y_encoded, epochs=50, batch_size=1)

#결과 출력
print("\n Accuracy: %.4f" % (model.evaluate(X, Y_encoded)[1]))
```
